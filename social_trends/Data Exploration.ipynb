```python
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
py2 = TrendReq(hl='en-US', tz=360)
py3 = TrendReq(hl='en-US', tz=360)
kw_list = ["Coronavirus"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 3-m', geo='US', gprop='')
py3.build_payload(kw_list, cat=0, timeframe='now 7-d', geo='US', gprop='')
py2.build_payload(kw_list, cat=0, timeframe='today 3-m', geo='US-MD', gprop='')
```

```python
df = pytrends.interest_over_time()
above = df[df["Coronavirus"] > 50]
above
```

```python
py2.interest_over_time()
```

```python
pytrends.interest_by_region(resolution='DMA', inc_low_vol=True, inc_geo_code=False)
```

```python
gtReg = py3.interest_by_region(resolution='REGION', inc_low_vol=True, inc_geo_code=False)
gtReg
```

```python
py2.interest_by_region(resolution='CITY', inc_low_vol=True, inc_geo_code=True)
```
